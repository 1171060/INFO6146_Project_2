```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, roc_curve
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from sklearn.neighbors import LocalOutlierFactor
```

```python
#download data
!gdown 1aRgPerDm858-r7zwAvaq2x2O3nits6M5
#read data
data  = pd.read_csv('/content/creditcard.csv')
data
```

```python
# Drop unnecessary columns and scale the data
scaler = StandardScaler()
all_data = data.drop(['Class', 'Time'], axis=1)
scaled_data = scaler.fit_transform(all_data)
```

```python
# Define the labels
labels = data['Class']
```

**Isolation Forest**

```python
# 1. Isolation Forest
iso_forest = IsolationForest(contamination=0.01, random_state=42)
iso_forest.fit(scaled_data)
iso_preds = iso_forest.predict(scaled_data)

# Convert Isolation Forest predictions to 1 (normal) and 0 (fraud)
iso_preds = [0 if x == -1 else 1 for x in iso_preds]
```

```python
# Evaluate Isolation Forest
print("=== Isolation Forest ===")
print(classification_report(labels, iso_preds))
iso_roc_auc = roc_auc_score(labels, iso_preds)
print(f"ROC AUC Score: {iso_roc_auc}")
```

```python
# Confusion Matrix for Isolation Forest
iso_conf_mat = confusion_matrix(labels, iso_preds)
sns.heatmap(iso_conf_mat, annot=True, fmt='d', cmap='Blues')
plt.title('Isolation Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

**Local Outlier Factor**

```python
# 2. Local Outlier Factor (LOF)
print("\n=== Local Outlier Factor (LOF) ===")
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.01)
lof_preds = lof.fit_predict(scaled_data)

# Convert LOF predictions to 1 (normal) and 0 (fraud)
lof_preds = [0 if x == -1 else 1 for x in lof_preds]
```

```python
# Evaluate Local Outlier Factor
print("\n=== Local Outlier Factor (LOF) ===")
print(classification_report(labels, lof_preds))
lof_roc_auc = roc_auc_score(labels, lof_preds)
print(f"ROC AUC Score: {lof_roc_auc}")
```

```python
# Confusion Matrix for LOF
lof_conf_mat = confusion_matrix(labels, lof_preds)
sns.heatmap(lof_conf_mat, annot=True, fmt='d', cmap='Blues')
plt.title('Local Outlier Factor Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
